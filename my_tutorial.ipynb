```python
coffee = pd.read_csv('./warmup-data/coffee.csv')
coffee
```

```python
#Modify data, change values in the day at index 0 & 1
coffee.loc[0, "Units Sold"] = 30
coffee.loc[1, "Units Sold"] = 50
coffee.head()
```

```python
#Set Multiple values /modify 
coffee.loc[0:3, "Units Sold"] = 20
coffee.head()
```

```python
#Sort values by column names,
coffee.sort_values("Units Sold", ascending=False)
```

```python
#Sort by multiple values
coffee.sort_values(["Units Sold", "Coffee Type"], ascending=[0,1])
```

```python
#iterate row by row
for index , row in coffee.iterrows():
  print (index)
  print (row['Coffee Type'])
  print("\n")
```
